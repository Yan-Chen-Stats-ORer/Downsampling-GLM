```python
import matplotlib.pyplot as plt
import numpy as np 
from UCI_utils_logit import *
from collections import Counter
from imblearn.datasets import fetch_datasets
from sklearn.model_selection import train_test_split
```

```python
yeast_me2 = fetch_datasets()['yeast_me2']
yeast_me2_df_feature = yeast_me2.data
yeast_me2_df_target = yeast_me2.target
#turn all the -1 to 0
yeast_me2_df_target[yeast_me2_df_target == -1] = 0
# print(pen_df_target)
sum(yeast_me2_df_target)/len(yeast_me2_df_target)
```

```python
yeast_me2_alpha_list = [0.032,0.033,0.034,0.035,0.036,0.037]
yeast_me2_log_loss_inv,yeast_me2_log_loss_inv_std,\
    yeast_me2_auc_inv,yeast_me2_auc_inv_std,\
    yeast_me2_log_loss_joint,yeast_me2_log_loss_joint_std,\
        yeast_me2_auc_joint,yeast_me2_auc_joint_std\
          = experiment_logit(np.log(1/(sum(yeast_me2_df_target)/len(yeast_me2_df_target))-1),yeast_me2_alpha_list,
                                 yeast_me2_df_feature,
                                 yeast_me2_df_target,exp_num=500)

#histogram 
plot_log_loss(yeast_me2_log_loss_inv,
              yeast_me2_log_loss_inv_std,
              yeast_me2_log_loss_joint,
              yeast_me2_log_loss_joint_std,
              yeast_me2_alpha_list, 
              sum(yeast_me2_df_target)/len(yeast_me2_df_target), 
              'yeast_me2',
              exp_num=500)
```
